```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
d = pd.read_csv('SMSSpamCollection',sep='\t',names=['Status','Message'])
print(d.head());
print(len(d[d.Status=='spam']))
len((d[d.Status=='ham']))
```

```python
x=d.Message
y=d.Status
d.loc[d["Status"]=='ham',"Status"]=1
d.loc[d["Status"]=='spam',"Status"]=0
d
```

```python
c=CountVectorizer()
x_c=c.fit_transform(["my name is pushpneet pushpneet singh","i am currently taking a session am"])
x_c.toarray()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
c1=CountVectorizer()
x_trainc1=c1.fit_transform(x_train)
x_trainc1.toarray()
```

```python
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
m=MultinomialNB()
m.fit(x_trainc1,y_train)

```

```python
x_testc1=c1.transform(x_test)
x_testc1.toarray()
```

```python
pre=m.predict(x_testc1)
pre
```

```python
print("Multinomial : {}".format(m.score(x_testc1,y_test)))
```
